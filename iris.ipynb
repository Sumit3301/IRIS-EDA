## Importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

## Importing the data

```python
df = pd.read_csv('IRIS.csv')
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df=df.dropna()  #dropping NaN or nll values if they existed
df
```

```python
plt.hist(df["species"],color="red")
print(df["species"].value_counts())
```

```python
#getting average length of petal length species wise
temp=pd.DataFrame(df)
df_avg=temp.groupby("species").mean()
print(df_avg)
```

```python
plt.bar(df_avg.index,df_avg["sepal_length"])
plt.xlabel("Species Type")
plt.ylabel("petal length")
plt.show()
```

```python
sns.pairplot(df, hue="species", diag_kind="hist")
```
